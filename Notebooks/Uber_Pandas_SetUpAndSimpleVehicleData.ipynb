```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Read file to pandas dataframe
df = pd.read_csv("uber.csv")

#print df.head to confirm dataframe
df.head(10)
```

```python
#Get high level summary of table
df.describe()
```

```python
#Drop rows with null values (5)
df.dropna(inplace = True)
```

```python
#Re-summarize table with null values omitted
df.describe()
```

```python
#Get activity counts per unique vehicles
df["vehicle_alias"].value_counts()
#Discovered need for name normalization
```

```python
#Create new df for just vehicle/activity data
vehicle_df = df["vehicle_alias"].value_counts()

#Define columns
vehicle_df.columns=['Vehicle', 'Uses']

#Create pie chart with vehicle data
plt.pie(vehicle_df, autopct="%1.2f%%", shadow=True, startangle=-90, radius=3)
```
